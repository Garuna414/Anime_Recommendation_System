```python
def convert_tags(obj):
    L = []
    i = obj.split(', ')
    L = L + i
    return L

def convert_desc(obj):
    L = []
    i = obj.split(' ')
    L = L + i
    return L

def convert_staffva(obj):
    L = []
    split_obj = obj.split(', ')
    colon = ' : '
    for i in split_obj:
        if colon in i:
            name = i.split(' : ')[1].strip()
            L.append(name)
        else:
            name = i.split('\n')[0].strip()
            L.append(name)
    return L

def convert_staff(obj):
    L = []
    dir = 'Director'
    crt = 'Creator'
    split_obj = obj.split(', ')
    for i in split_obj:
        name = i.split(' : ')
        if dir in name[1]:
            L.append(name[0])
        elif crt in name[1]:
            L.append(name[0])
    return L
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
anime_tv['Synopsis'] = anime_tv['Synopsis'].apply(stem)
anime_movies['Synopsis'] = anime_movies['Synopsis'].apply(stem)
```
